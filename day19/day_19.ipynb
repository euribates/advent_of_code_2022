However, by using blueprint 2 in the example above, you could do even better: the largest number of geodes you could open in 24 minutes is **12**.

Determine the quality level of each blueprint by multiplying that blueprint's ID number with the largest number of geodes that can be opened in 24 minutes using that blueprint. In this example, the first blueprint has ID 1 and can open 9 geodes, so its quality level is 9. The second blueprint has ID 2 and can open 12 geodes, so its quality level is 24. Finally, if you add up the quality levels of all of the blueprints in the list, you get 33.

Determine the quality level of each blueprint using the largest number of geodes it could produce in 24 minutes. What do you get if you add up the quality level of all of the blueprints in your list?

````python
from enum import Enum
import collections
import re

import pytest

from icecream import ic
````

````python
def load_input(filename):
    with open(filename, 'r') as f_input:
        for line in f_input:
            line = line.strip()
            yield line

ORE = 'ORE'
CLAY = 'CLAY'
OBSIDIAN = 'OBSIDIAN'
GEODE = 'GEODE'

Material = Enum('Material', [ORE, CLAY, OBSIDIAN, GEODE])
            
class Robot:
    
    def __init__(self, produces: Material):
        self.produces = produces
        self.counter = 0
        self.costs = {}
        
    def __str__(self):
        return f"{self.produces}-collecting Robot"
    
    def add_cost(self, material, units):
        self.costs[material] = units
        
    def run(self, inventory):
        self.counter += 1
        return (1, self.produces)
    
    def copy(self):
        new_robot = Robot(produces=self.produces)
        new_robot.costs = self.costs.copy()
        return new_robot
    
    def can_build(self, inventory):
        return all([
            material in inventory and inventory[material] >= cost
            for material, cost in self.costs.items()
        ])
        
    def build(self, inventory):
        for material, cost in self.costs.items():
            inventory[material] -= cost
        return self.copy(), inventory

        
def create_ore_robot(rule: str):
    pat_ore_robot = re.compile(r'Each ore robot costs (\d+) ore.')
    _match = pat_ore_robot.match(rule)
    robot = Robot(produces=Material.ORE)
    robot.add_cost(Material.ORE, int(_match.group(1)))
    return robot

    
ore_robot = create_ore_robot('Each ore robot costs 4 ore.')
assert ore_robot.produces == Material.ORE
assert ore_robot.costs[Material.ORE] == 4
assert ore_robot.counter == 0

robots = [ore_robot]
costs = collections.defaultdict(int)
inventory = collections.defaultdict(int)
     
    
def show_robots(robots):
    buff = []
    ore_robots = sum([1 for r in robots if r.produces == Material.ORE])
    if ore_robots: 
        buff.append(f"{ore_robots} ore-collecting robots")

    clay_robots = sum([1 for r in robots if r.produces == Material.CLAY])
    if clay_robots: 
        buff.append(f"{clay_robots} clay-collecting robots")
                           
    obsidian_robots = sum([1 for r in robots if r.produces == Material.OBSIDIAN])
    if obsidian_robots: 
        buff.append(f"{obsidian_robots} obsidian-collecting robots")

    geode_robots = sum([1 for r in robots if r.produces == Material.GEODE])
    if geode_robots: 
        buff.append(f"{geode_robots} geode-cracking robots")
    return ','.join(buff)
    
    
def show_inventory(inventory):
    ore = inventory[Material.ORE]
    clay = inventory[Material.CLAY]
    obsidian = inventory[Material.OBSIDIAN] 
    geode = inventory[Material.GEODE] 
    return f"[ ORE: {ore:8} | CLAY: {clay:8} | OBSIDIAN: {obsidian:8} | GEODE: {geode:8} ]"

for _tick in range(6):
    print(_tick)
    for robot in robots:
        print('\t- ', robot, 'produce 1', robot.produces, 'unit')
        inventory[robot.produces] += 1
    if robot.can_build(inventory):
        new_robot, inventory = robot.build(inventory)
        robots.append(new_robot)
    print('\t- ', show_robots(robots), show_inventory(inventory))
        
            
````

````python
def create_clay_robot(rule: str):
    pat_clay_robot = re.compile(r'Each clay robot costs (\d+) ore.')
    _match = pat_clay_robot.match(rule)
    robot = Robot(produces=Material.CLAY)
    robot.add_cost(Material.ORE, int(_match.group(1)))
    return robot

clay_robot = create_clay_robot('Each clay robot costs 2 ore.')
assert clay_robot.produces == Material.CLAY
assert clay_robot.costs[Material.ORE] == 2
````

````python
def create_obsidian_robot(rule: str):
    pat_obsidian_robot = re.compile(r'Each obsidian robot costs (\d+) ore and (\d+) clay.')
    _match = pat_obsidian_robot.match(rule)
    robot = Robot(produces=Material.OBSIDIAN)
    robot.add_cost(Material.ORE, int(_match.group(1)))
    robot.add_cost(Material.CLAY, int(_match.group(2)))  
    return robot

obsidian_robot = create_obsidian_robot('Each obsidian robot costs 3 ore and 14 clay.')
assert obsidian_robot.produces == Material.OBSIDIAN
assert obsidian_robot.costs[Material.ORE] == 3
assert obsidian_robot.costs[Material.CLAY] == 14
````

````python
def create_geode_robot(rule: str):
    pat_geode_robot = re.compile(r'Each geode robot costs (\d+) ore and (\d+) obsidian.')
    _match = pat_geode_robot.match(rule)
    robot = Robot(produces=Material.GEODE)
    robot.add_cost(Material.ORE, int(_match.group(1)))
    robot.add_cost(Material.OBSIDIAN, int(_match.group(2)))
    return robot

geode_robot = create_geode_robot('Each geode robot costs 6 ore and 11 obsidian.')
assert geode_robot.produces == Material.GEODE
assert geode_robot.costs[Material.ORE] == 6
assert geode_robot.costs[Material.OBSIDIAN] == 11
````
